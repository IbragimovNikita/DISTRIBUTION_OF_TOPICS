### Распределение статей

```python
import re
```

```python
# Читаем файл и заносим его в список построчно без разделителей
with open(r'C:\Users\serge\Desktop\Data Science\DZ1\urls.txt') as file:
    list_themes = file.read().splitlines()
print(list_themes)
```

```python
# Удаляем слеши и оставляем только все темы
for i in range(len(list_themes)):
    list_themes[i] = re.findall(r'^/\w+/', list_themes[i]) # оставляем только темы
list_themes = [x for l in list_themes for x in l] # убираем вложенность
list_themes = [x.split('/')[1] for x in list_themes]
print(list_themes)
```

```python
# Создаем сет, где хранятся уникальные названия статей
set_themes = set(list_themes)
print(set_themes)
```

```python
def counter_themes(list_themes, in_theme):
    counter = 0
    for theme in list_themes:
        if theme == in_theme:
            counter += 1
    return in_theme, counter

ready_tuple_themes = []

for themes in set_themes:
    ready_tuple_themes.append(counter_themes(list_themes, themes))
print(ready_tuple_themes)
```

```python
ready_dict_themes = dict(ready_tuple_themes)
ready_dict_themes
```
